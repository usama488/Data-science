```python
from google.colab import files
import io
import pandas as pd # Ensure pandas is imported here or in a previous cell

print("Please select your CSV file from your desktop:")
uploaded = files.upload()

for fn in uploaded.keys():
  print(f'User uploaded file "{fn}" with length {len(uploaded[fn])} bytes')
  # Assuming the uploaded file is your dataset, read it into a DataFrame
  try:
    df = pd.read_csv(io.BytesIO(uploaded[fn]))
    print("Dataset loaded successfully!")
    display(df.head())
  except Exception as e:
    print(f"Error reading file {fn}: {e}")
```

```python
df_cleaned = df.copy()

# Convert 'Size' column to numeric
def clean_size(size):
    if 'M' in str(size):
        return float(str(size).replace('M', '')) * 1000000
    elif 'k' in str(size):
        return float(str(size).replace('k', '')) * 1000
    return pd.NA
df_cleaned['Size'] = df_cleaned['Size'].apply(clean_size)
df_cleaned['Size'] = pd.to_numeric(df_cleaned['Size'])

# Convert 'Installs' column to numeric, handling 'Free'
df_cleaned['Installs'] = df_cleaned['Installs'].str.replace('+', '', regex=False).str.replace(',', '', regex=False).str.replace('Free', '0', regex=False).astype(int)

# Convert 'Price' column to numeric, handling 'Free' and 'Everyone'
df_cleaned['Price'] = df_cleaned['Price'].str.replace('$', '', regex=False).str.replace('Free', '0', regex=False).str.replace('Everyone', '0', regex=False).astype(float)

print("Data types after conversion:")
df_cleaned[['Size', 'Installs', 'Price']].info()
```
