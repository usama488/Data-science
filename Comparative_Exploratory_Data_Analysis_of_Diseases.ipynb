# Comparative Exploratory Data Analysis (EDA) of Diseases

## Objective
To compare two disease datasets — Breast Cancer and Parkinson’s Disease — using exploratory data analysis to understand differences in data structure, feature behavior, and biological relevance without performing any prediction.


## Step 1: Import Required Libraries
Libraries used for data manipulation and visualization.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

## Step 2: Load Datasets
Load Breast Cancer and Parkinson’s Disease datasets.


```python
# Breast Cancer Dataset
df = pd.read_csv("/content/data.csv")
```

```python
df.head()
```

```python
# Parkinson's Dataset
parkinsons_df = pd.read_csv("/content/parkinsons.data")
```

```python
df.head()
```

## Step 3: Data Cleaning
Remove unnecessary identifier columns and check missing values.


```python
# Breast Cancer cleaning
cancer_df.drop(columns=['id', 'Unnamed: 32'], inplace=True)

# Parkinson’s cleaning
parkinsons_df.drop(columns=['name'], inplace=True)

print(cancer_df.isnull().sum().sum())
print(parkinsons_df.isnull().sum().sum())
```

## Step 4: Dataset Overview
Compare dataset size and feature counts.


```python
overview = pd.DataFrame({
    "Dataset": ["Breast Cancer", "Parkinson’s"],
    "Rows": [cancer_df.shape[0], parkinsons_df.shape[0]],
    "Columns": [cancer_df.shape[1], parkinsons_df.shape[1]]
})

overview
```

## Step 5: Target Variable Distribution
Compare disease vs healthy class distribution.


```python
fig, axes = plt.subplots(1, 2, figsize=(12,5))

sns.countplot(x='diagnosis', data=cancer_df, ax=axes[0])
axes[0].set_title("Breast Cancer Diagnosis Distribution")

sns.countplot(x='status', data=parkinsons_df, ax=axes[1])
axes[1].set_title("Parkinson’s Status Distribution")

plt.show()
```

## Step 6: Feature Distribution Comparison
Compare numeric feature distributions from both datasets.


```python
fig, axes = plt.subplots(1, 2, figsize=(12,5))

sns.histplot(cancer_df['radius_mean'], bins=30, ax=axes[0])
axes[0].set_title("Breast Cancer: Radius Mean")

sns.histplot(parkinsons_df['MDVP:Fo(Hz)'], bins=30, ax=axes[1])
axes[1].set_title("Parkinson’s: Fundamental Frequency")

plt.show()
```

## Step 7: Disease-wise Feature Comparison
Analyze how a key feature behaves across disease classes.


```python
fig, axes = plt.subplots(1, 2, figsize=(12,5))

sns.boxplot(x='diagnosis', y='radius_mean', data=cancer_df, ax=axes[0])
axes[0].set_title("Radius Mean by Cancer Diagnosis")

sns.boxplot(x='status', y='MDVP:Fo(Hz)', data=parkinsons_df, ax=axes[1])
axes[1].set_title("Voice Frequency by Parkinson’s Status")

plt.show()
```

## Step 8: Correlation Analysis
Visualize correlation structure of both datasets.


```python
fig, axes = plt.subplots(1, 2, figsize=(16,6))

sns.heatmap(cancer_df.drop('diagnosis', axis=1).corr(), cmap='coolwarm', ax=axes[0])
axes[0].set_title("Breast Cancer Feature Correlation")

sns.heatmap(parkinsons_df.corr(), cmap='coolwarm', ax=axes[1])
axes[1].set_title("Parkinson’s Feature Correlation")

plt.show()
```

## Step 9: Comparative Insights

### Data Nature
- Breast Cancer data contains clinical tumor measurements.
- Parkinson’s data is based on biomedical voice signals.

### Feature Behavior
- Cancer features show strong separation based on tumor size.
- Parkinson’s features show subtle variability in frequency and amplitude.

### Biological Relevance
- Tumor size and shape are clinically linked to malignancy.
- Voice frequency instability reflects neuromuscular impairment.

### Exploratory Value
Both datasets show meaningful biological patterns suitable for EDA without predictive modeling.


## Conclusion
Comparative EDA highlights how different diseases require different analytical perspectives.
Clinical measurements and signal-based biomarkers both provide valuable insights when explored systematically.


The comparative exploratory data analysis of the Breast Cancer and Parkinson's Disease datasets has led to several key insights:

Data Nature

The Breast Cancer data primarily consists of clinical tumor measurements, providing quantitative descriptions of cell nuclei features.
The Parkinson’s data, on the other hand, is derived from biomedical voice signals, capturing acoustic properties that reflect vocal cord and respiratory system function.

Feature Behavior

For Breast Cancer, features like radius_mean show strong differentiation between malignant and benign cases, indicating the importance of tumor size and morphology.

In the Parkinson's dataset,

 features related to voice frequency and amplitude (MDVP:Fo(Hz)) show subtle variabilities that are indicative of neuromuscular impairment.

Biological Relevance

The clinical measurements in the Breast Cancer dataset (e.g., tumor size, perimeter, concavity) are directly linked to the biological processes of malignancy.
The voice signal instability in Parkinson's reflects underlying neurological and muscular issues affecting speech production.

Exploratory Value


Both datasets, despite their different origins and data types, exhibit meaningful biological patterns that are amenable to exploratory data analysis without needing immediate predictive modeling. The EDA helped in understanding data structures, feature distributions, and inter-feature relationships (correlations) within each disease context.

In conclusion,

 this comparative EDA highlights that different diseases often require distinct analytical perspectives. While clinical measurements provide direct insights for diseases like breast cancer, signal-based biomarkers are crucial for conditions like Parkinson's. Both approaches, when explored systematically, yield valuable insights into the diseases.
