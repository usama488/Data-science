# Exploratory Data Analysis (EDA) on Heart Disease Dataset

This Exploratory Data Analysis (EDA) focuses on examining patient clinical data related to heart disease. The analysis explores key attributes such as age, gender, cholesterol level, blood pressure, chest pain type, and other medical factors to identify patterns and relationships associated with heart disease, without performing any prediction or modeling.

# Libraries Import

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# handling dataset

# Disease Distribution

# Kitne patients diseased hain aur kitne healthy.

```python
df['target'].value_counts()
```

# Univariate Analysis

```python
df['age'].hist()
plt.title("Age Distribution")
plt.show()
```

# Relationship Analysis

```python
sns.boxplot(x='target', y='age', data=df)
plt.show()
```

# Gender vs Disease

```python
pd.crosstab(df['sex'], df['target'])
```

# Females sex=0: 24 individuals do not have heart disease, while 72 individuals do.
# Males sex=1: 114 individuals do not have heart disease, while 92 individuals do.

# Correlation Analysis

```python
sns.heatmap(df.corr(), annot=True)
plt.show()
```

# EDA Insights


# Disease Distribution: The dataset now contains 164 patients with heart disease (target=1) and 138 patients without heart disease (target=0).

## Gender vs. Heart Disease: A crosstabulation revealed the following:
# Among females (sex=0): 24 individuals do not have heart disease, while 72 individuals do. This suggests a higher proportion of heart disease among females in this dataset.
# Among males (sex=1): 114 individuals do not have heart disease, while 92 individuals do.
