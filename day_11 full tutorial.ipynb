# Lets select only those rows which have paod more than 200 pounds

```python
import pandas as pd
import seaborn as sns

# Re-load df and preprocess necessary columns
df = sns.load_dataset('titanic')

# Fill missing age values, as 'age_groups' depends on 'age'
df['age'] = df['age'].fillna(df['age'].mean())

# Define bins and labels for 'age_groups'
bins=[0,1,5,12,18,30,50,80]
labels=['infant','toddler','kids','teens','younger','middle aged','old']

# Create 'binned_age' and rename to 'age_groups'
df['binned_age'] = pd.cut(df['age'],bins=bins,labels=labels)
df.rename(columns={'binned_age':'age_groups'},inplace=True)

# Select relevant columns into df_01 and filter
df_01=df[['survived','age_groups','fare','class']]
print(df_01[df_01['fare']>200])
```

```python
df_200 = df_01[df_01['fare'] > 200]
print(df_200.head())
```
